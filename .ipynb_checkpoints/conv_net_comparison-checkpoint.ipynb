## Importing Necessary Tools

```python
import os
import glob
import imageio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.utils import img_to_array, load_img

import keras
from keras.preprocessing import image as keras_image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model, Sequential
from keras.layers import (
    GlobalMaxPooling2D,
    Conv2D,
    MaxPooling2D,
    Flatten,
    Bidirectional,
    SpatialDropout2D,
    Input,
    Dropout,
    Dense,
    Activation,
    GlobalAveragePooling2D,
)
from keras.applications import imagenet_utils, MobileNet
from keras.applications.mobilenet import preprocess_input
from keras.optimizers import Adam
from keras.utils.vis_utils import plot_model
from keras.utils.vis_utils import model_to_dot


from sklearn.svm import SVC

from IPython.display import Image, display
from splitfolders import split

import pygraphviz as pgv
```

```python
def prepare_image(file):
    # Load image and resize it to the target size (224x224)
    img = load_img(file, target_size=(224, 224))
    
    # Convert the image to a NumPy array
    img_array = img_to_array(img)
    
    # Expand the dimensions of the array to match the expected input shape for MobileNet
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    
    # Preprocess the input image
    return keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

## Convolutional Neural Network (CNN) Modeling

### Data Splitting

```python
splitfolders.ratio("Soda Bottles", output="TVT_split", seed=1337, ratio=(0.8, 0.1, 0.1))
```

### Generating Data Sets

```python
# Set the batch size for the data generator
batch_size = 300
```

```python
# Define data generators for training and testing with specified augmentations
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    # rotation_range=5,  # Commented out rotation augmentation
    width_shift_range=0.2,
    height_shift_range=0.2,
)

test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)  # No augmentation for the validation and test datasets
```

```python
# Create a training data generator with augmentation
train_data = train_datagen.flow_from_directory(
    "TVT_split/train",  # Path to the training dataset
    target_size=(224, 224),  # Resize images to (224, 224)
    color_mode="rgb",  # Use RGB color mode
    batch_size=batch_size,  # Set the batch size
    class_mode="categorical",  # Use categorical labels for multiple classes
)
```

```python
# Create a validation data generator without augmentation
validation_data = test_datagen.flow_from_directory(
    "TVT_split/val",  # Path to the validation dataset
    target_size=(224, 224),  # Resize images to (224, 224)
    batch_size=batch_size,  # Set the batch size
    color_mode="rgb",  # Use RGB color mode
    class_mode="categorical",  # Use categorical labels for multiple classes
)
```

```python
# Create a test data generator without augmentation
test_data = test_datagen.flow_from_directory(
    "TVT_split/test",  # Path to the test dataset
    target_size=(224, 224),  # Resize images to (224, 224)
    batch_size=batch_size,  # Set the batch size
    color_mode="rgb",  # Use RGB color mode
    class_mode="categorical",  # Use categorical labels for multiple classes
)
```

### Designing a ConvNet for Image Classification From Scratch

Check out the following link for more details: [Building Powerful Image Classification Models Using Very Little Data](https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html)

```python
# Define a Sequential model
model1 = Sequential()

# Convolutional Layer 1 for Feature Extraction
model1.add(Conv2D(32, kernel_size=(3, 3), input_shape=(224, 224, 3)))
model1.add(Activation("relu"))  # Introduce non-linearity
model1.add(MaxPooling2D(pool_size=(2, 2)))  # Feature selection through max-pooling
model1.add(Dropout(0.2))  # Regularization to prevent overfitting

# Convolutional Layer 2 for Feature Extraction
model1.add(Conv2D(32, (3, 3)))
model1.add(Activation("relu"))
model1.add(MaxPooling2D(pool_size=(2, 2)))
model1.add(Dropout(0.2))

# Convolutional Layer 3 for Feature Extraction
model1.add(Conv2D(64, (3, 3)))
model1.add(Activation("relu"))
model1.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten Layer to Convert 3D Feature Maps to 1D Feature Vectors
model1.add(Flatten())

# Dense Layer 1 for Flexibility
model1.add(Dense(128))
model1.add(Activation("relu"))
model1.add(Dropout(0.2))  # Regularization

# Dense Layer 2 for Final Output (8 categories)
model1.add(Dense(8))
model1.add(Activation("softmax"))  # Activation for multi-class classification

# Model Compilation
model1.compile(optimizer="Adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Display model summary
model1.summary()
```

## Transfer Learning

### Designing a Transfer Learning Approach for Image Classification

Check out the following link for more details: [Pre-Trained Models for Transfer Learning in Keras](https://www.educative.io/answers/pre-trained-models-for-transfer-learning-in-keras)


```python
# Load the Xception pre-trained model from Keras Applications
# - weights: "imagenet" initializes the model with weights pre-trained on ImageNet data
# - include_top: False excludes the top (fully connected) layers of the model, making it suitable for transfer learning
# - input_shape: Specifies the input shape of the images the model will process
pretrained_model = tf.keras.applications.Xception(
    weights="imagenet", include_top=False, input_shape=(224, 224, 3)
)
```

```python
# Set the pre-trained part of the model to be non-trainable
pretrained_model.trainable = False
```

```python
# Define the second model with the pre-trained Xception base
model2 = Sequential(pretrained_model)

# Add layers to the model
model2.add(GlobalAveragePooling2D())  # Average pooling for global spatial average
model2.add(Dense(1024, activation="relu"))  # Dense layer with ReLU activation
model2.add(Dropout(0.2))  # Dropout for regularization
model2.add(Dense(512, activation="relu"))  # Dense layer with ReLU activation
model2.add(Dropout(0.2))  # Dropout for regularization
model2.add(Dense(256, activation="relu"))  # Dense layer with ReLU activation
model2.add(Dense(8, activation="softmax"))  # Output layer with softmax activation for 8 classes

# Model Compilation
model2.compile(optimizer="Adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Display model summary
model2.summary()
```

```python
# Visualize the architecture of the second model
graph2 = model_to_dot(model2, show_shapes=True, show_layer_names=True, rankdir="TB", expand_nested=True)

# Render the graph to a PNG file
graph2.write_png('model2.png')

# Display the generated image
Image(filename='model2.png')
```
